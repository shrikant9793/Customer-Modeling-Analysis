```python
#categorical Features
list_1 = list(data.columns)
```

```python
list_1
```

```python
list_cate= []
for i in list_1:
    if data[i].dtype== 'object':
        list_cate.append(i)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in list_cate:
    data[i]=le.fit_transform(data[i])
```

```python
data
```

```python
X = data.drop('Geography', axis= 1)
y = data['Geography']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense


classifier = Sequential()

classifier.add(Dense(activation= 'relu', units= 6, kernel_initializer= 'uniform'))#1st hidden layer

classifier.add(Dense(activation= 'relu', input_dim= 11, units= 6, kernel_initializer= "uniform"))# 2nd hidden layer

classifier.add(Dense(activation = 'sigmoid', units= 1, kernel_initializer = 'uniform'))#output layer

classifier.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'])#compile the ANN

classifier.fit(X_train, y_train, batch_size= 10,
              epochs= 200,
              )#fitting the ANN to the training set
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
print(y_pred > 0.5)
```
